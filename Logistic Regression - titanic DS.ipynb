```python
Xtrain=df.iloc[:,0:6]
Ytrain=df.iloc[:,6]
print(Xtrain.shape)
Xtest=df1.iloc[:,:]
Xtrain=Xtrain.to_numpy()
print(Xtest.shape)
Xtest=Xtest.to_numpy()
```

```python
 for i in range(6):
        for j in range(i,6):
            a=[[0 for j in range(1)] for i in range(Xtrain.shape[0])]
            for k in range(Xtrain.shape[0]):
                a[k][0]=Xtrain[k][i]*Xtrain[k][j]
            Xtrain=np.append(Xtrain,a,axis=1)
```

```python
for i in range(6):
        for j in range(i,6): 
            h=[ [0 for j in range(1)] for i in range(Xtest.shape[0])]
            for k in range(Xtest.shape[0]):
                 h[k][0]=Xtest[k][i]*Xtest[k][j]
            
            Xtest=np.append(Xtest,h,axis=1)
```

```python
scaler= preprocessing.MinMaxScaler(feature_range=(0,100))
scaler.fit(Xtrain)
Xtrain=scaler.transform(Xtrain)
```
